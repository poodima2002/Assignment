```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()

# Read image
img = cv.imread(list(uploaded.keys())[0], 0)

# Apply Canny
edges = cv.Canny(img, 550, 690)

# Display
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(edges, cmap='gray')
plt.title("Canny Edges")

plt.show()
```

```python
# Extract coordinates
indices = np.where(edges != 0)
x = indices[1]
y = indices[0]

# Plot
plt.figure(figsize=(6,6))
plt.scatter(x, y, s=2)

plt.gca().invert_yaxis()

plt.title("Scatter Plot of Edge Points")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```python
# Least Squares
X = np.column_stack((x, np.ones(len(x))))
m, c = np.linalg.lstsq(X, y, rcond=None)[0]

# Line
line_x = np.linspace(np.min(x), np.max(x), 100)
line_y = m * line_x + c

# Plot
plt.figure(figsize=(6,6))
plt.scatter(x, y, s=2)
plt.plot(line_x, line_y, 'r')

plt.gca().invert_yaxis()
plt.title("Least Squares Fit")

plt.show()

print("Slope:", m)
print("Intercept:", c)
```

```python
# TLS
x_mean, y_mean = np.mean(x), np.mean(y)

U = np.column_stack((x - x_mean, y - y_mean))
UTU = np.dot(U.T, U)

eigenvalues, eigenvectors = np.linalg.eigh(UTU)

a, b = eigenvectors[:, np.argmin(eigenvalues)]

m_tls = -a / b
c_tls = y_mean - m_tls * x_mean

line_y_tls = m_tls * line_x + c_tls

# Plot
plt.figure(figsize=(6,6))
plt.scatter(x, y, s=2)
plt.plot(line_x, line_y_tls, 'b')

plt.gca().invert_yaxis()
plt.title("Total Least Squares")

plt.show()
```

```python
theta_tls = np.degrees(np.arctan(m_tls))
print(f"Angle (TLS): {theta_tls:.2f} degrees")
```

```python
from sklearn.linear_model import RANSACRegressor

ransac = RANSACRegressor()
ransac.fit(x.reshape(-1,1), y)

m_r = ransac.estimator_.coef_[0]
c_r = ransac.estimator_.intercept_

line_y_r = m_r * line_x + c_r

# Plot
plt.figure(figsize=(6,6))
plt.scatter(x, y, s=2)
plt.plot(line_x, line_y_r, 'g')

plt.gca().invert_yaxis()
plt.title("RANSAC Fit")

plt.show()
```
